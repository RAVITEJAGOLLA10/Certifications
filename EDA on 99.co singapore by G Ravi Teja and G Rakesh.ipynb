```python
title_class = '_3Ajbv _30I97 _1vzK2'
tag = 'a'
url = 'https://www.99.co/singapore/sale'
page = requests.get(url,headers = {'User-Agent': 'Mozilla/5.0'}) # 1,
if page.status_code==200:
    soup = BeautifulSoup(page.text) # 2
    title = soup.find_all(tag, attrs={'class':title_class}) # 3
    title = [x.text for x in title]
    print(title)
else:
    print("Http Error")
```

```python
name = [] 
address = []
typeofhouse = []
yearofmanufac = []
price = []
offer = []
bedbathsizeinfo = []
pagenum = []
# go throught the 21 pages
total_time = time.time()
for i in range(1,18): 
    start_time = time.time()
    URL = "https://www.99.co/singapore/sale?bathrooms=any&composite_floor_level=any&composite_furnishing=any&composite_property_status=any&composite_views=any&features_and_amenities=any&has_floor_plan=false&isFilterUnapplied=false&main_category=all&map_bounds=1.5827095153768858%2C103.49449749970108%2C1.1090706240313446%2C104.12483807587296&page_num={}&page_size=35&path=%2Fsingapore%2Fsale&period_of_availability=any&property_segments=residential&query_coords=1.3039947%2C103.8298507&query_limit=radius&query_type=city&rooms=any&show_cluster_preview=true&show_description=true&show_future_mrts=true&show_internal_linking=true&show_meta_description=true&show_nearby=true&tenure=any&zoom=11".format(i)
    page = requests.get(URL,headers = {'User-Agent': 'Mozilla/5.0'})
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)

    for x in soup.find_all('div', attrs={'class':'TaGB0'}):
        nameofhouse = x.find('a', attrs = {'class':'_3Ajbv _30I97 _1vzK2'})
        streetaddress = x.find('li', attrs = {'class':'_3WG9R _3L5OV'})
        typeofhome = x.find('li', attrs = {'class':'_1LPAx'})
        year = x.find('span', attrs = {'class':'JiLXa'})
        priceofhouse = x.find('li', attrs={'class':'JlU_W'})
        offerper = x.find('li', attrs={'class':'_1iMmV'})
        bedbathsize = x.find('span', attrs={'class':'cUjn9'})
        if nameofhouse is None:
            name.append(np.NaN)
        else:
            name.append(nameofhouse.text)
            
        if streetaddress is None:
            address.append(np.NaN)
        else:
            address.append(streetaddress.text)
            
        if typeofhome is None:
            typeofhouse.append(np.NaN)
        else:
            typeofhouse.append(typeofhome.text)
            
        if year is None:
            yearofmanufac.append(np.NaN)
        else:
            yearofmanufac.append(year.text)
            
        if priceofhouse is None:
            price.append(np.NaN)
        else:
            price.append(priceofhouse.text)
            
        
        if offerper is None:
            offer.append(np.NaN)
        else:
            offer.append(offerper.text)
        
        if bedbathsize is None:
            bedbathsizeinfo.append(np.NaN)
        else:
            bedbathsizeinfo.append(bedbathsize.text)
        pagenum.append(i)
    print('Page {} completed in {} seconds'.format(i, time.time()-start_time))
print("Total Time Completed in seconds"+str(time.time()-total_time))
```
